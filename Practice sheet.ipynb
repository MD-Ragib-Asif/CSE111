```python
#task 1
class PlayerEarning:
    def __init__(self, name):
        self.name = name
        
    def calculateTotal(self, earning, goal=None):
        self.earning = earning
        self.goal = goal
        if isinstance(self.goal, int):
            if self.goal >= 30:
                self.bonus = (5/100)*self.earning + 10000
            else:
                self.bonus = (5/100)*self.earning
        else:
            self.bonus = 0
        self.total = self.earning + self.bonus
        
    def printDetails(self):
        print(f'Player Name: {self.name} \nPlayer Season Earning without bonus: {self.earning} \nBonus: {int(self.bonus)} \nPlayer Season Earning After Bonus: {int(self.total)}')


#====================================================================
print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()
print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()
print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
#task 2
class myList:
#     lst=[]
    def __init__(self, *args):
        self.lst=[]
        self.total = 0
        if len(args) == 0:
            self.lst.append(0)
        else:
            for i in args:
                self.lst.append(i)
    
    def merge(self, *args):
        for i in args:
            self.lst.append(i)
            
    def sum(self):
        total = 0
        for i in self.lst:
            total += i
        self.total = total
        print(f'Sum: {self.total}')
        
    def average(self):
        if self.total == 0:
            average = self.total
        else:
            average = self.total/len(self.lst)
        print(f'Average: {average}')


#=================================================================
l1 = myList(2,3,4,5,6) #you might need a list inside your class to store the values
l1.sum()
l1.merge(4,5,9)
l1.sum()
l1.average()
print("-----------------------------")
l2 = myList()
l2.average()
l2.merge(1,2,4,8)
l2.sum()
```

```python
#task 3
class Bird:
    def __init__(self, name, fly=False, type = 'Flightless Birds'):
        self.name = name
        self.fly_s = fly
        self.type = type
        
    def fly(self):
        if self.fly_s is True:
            print(f'{self.name} can fly')
        else:
            print(f'{self.name} can not fly')
            
    def setType(self, birdtype):
        self.type = birdtype
        
    def printDetail(self):
        print(f'Name: {self.name} \nType: {self.type}')

#==========================================================
ostrich = Bird('Ostrich')
duck = Bird("Duck", True)
owl = Bird('Owl', True)
print("###########################")
ostrich.fly()
duck.fly()
owl.fly()
duck.setType('Water Birds')
owl.setType('Birds of Prey')
print("=========================")
ostrich.printDetail()
print("=========================")
duck.printDetail()
print("=========================")
owl.printDetail()
```

```python
#task 4
class Account:
    count = 0
    def __init__(self, name, age, occupation, salary):
        self.name = name
        self.age = age
        self.occupation = occupation
        self.salary = salary
        Account.count += 1
        
    def addMoney(self, money):
        self.salary += money
        
    def withdrawMoney(self, money):
        if money <= self.salary:
            self.salary -= money
        
    def printDetails(self):
        print(f'Name: {self.name} \nAge: {self.age} \nOccupation: {self.occupation} \nTotal Amount: {self.salary}')


#=================================================================
print('No of account holders:', Account.count)
print("=========================")
p1 = Account("Abdul", 45, "Service Holder", 500000)
p1.addMoney(300000)
p1.printDetails()
print("=========================")
p2 = Account("Rahim", 55, "Businessman", 700000)
p2.withdrawMoney(700000)
p2.printDetails()
print("=========================")
p3 = Account("Ashraf", 62, "Govt. Officer", 200000)
p3.withdrawMoney(250000)
p3.printDetails()
print("=========================")
print('No of account holders:', Account.count)
```

```python
#task 5
class Smartphone:
    
    def __init__(self, name=None):
        self.name = name
        self.dic={}
        
    def addFeature(self, feature_type, feature_details):
        if self.name == None:
            print(f'Feature can not be added without phone name')
        else:
            if feature_type in self.dic:
                self.dic[feature_type].append(feature_details)
            else:
                self.dic[feature_type] = [feature_details]
            
            
    def setName(self, name):
        self.name = name
        
    def printDetail(self):
        string = ""
        for i in self.dic:
            string += f'{i}: {", ".join(self.dic[i])}\n'
        print(f'Phone Name: {self.name} \n' + string)
    

#======================================================================
s1 = Smartphone()
print("=================================")
s1.addFeature("Display", "6.1 inch")
print("=================================")
s1.setName("Samsung Note 20")
s1.addFeature("Display", "6.1 inch")
s1.printDetail()
print("=================================")
s2 = Smartphone("Iphone 12 Pro")
s2.addFeature("Display", "6.2 inch")
s2.addFeature("Ram", "6 GB")
print("=================================")
s2.printDetail()
s2.addFeature("Display", "Amoled panel")
s2.addFeature("Ram", "DDR5")
print("=================================")
s2.printDetail()
print("=================================")
```
