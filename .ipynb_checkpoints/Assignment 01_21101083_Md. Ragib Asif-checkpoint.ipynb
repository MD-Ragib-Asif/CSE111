```python
# task 1 (string)
N = input("Please enter a string: ")

countCap= 0
countSma= 0
for i in N:
    if 65<= ord(i) <=90:
        countCap += 1
    elif 97<= ord(i) <=122:
        countSma += 1
          
    
string=""
if countSma >= countCap:
    for j in N:
        if 65<= ord(j) <=90:
            string+=chr(ord(j)+32)
        elif 97<= ord(j) <=122:
            string+=j
else:
    for j in N:
        if 65<= ord(j) <=90:
            string+=j
        elif 97<= ord(j) <=122:
            string+=chr(ord(j)-32)

print(string)
```

```python
# task 2 (string)
N = input("Please enter a string: ")

countWord= 0
countNum= 0
for i in N:
    if (65<= ord(i) <=90) or (97<= ord(i) <=122):
        countWord += 1
    else:
        countNum += 1
        
        
if countWord==0:
    print("NUMBER")
elif countNum==0:
    print("WORD")
else:
    print("MIXED")
```

```python
# task 5 (string)
N1=input("Please enter first string: ")
N2=input("Please enter second string: ")

string=""
for i in N1:
    if i in N2:
        string+=i
        
for j in N2:
    if j in N1:
        string+=j
        
if len(string)!=0:
    print(string)
else:
    print("Nothing in common")
```

```python
# task 6 (string)
N = input("Please enter a valid password: ")

special=["_","$","#","@"]

lowCheck = False
upCheck = False
digCheck = False
speCheck = False

for i in N:
    if 65<= ord(i) <=90:
        upCheck = True
    if 97<= ord(i) <=122:
        lowCheck = True
    if 48<= ord(i) <=57:
        digCheck = True
    if i in special:
        speCheck = True

lst=[]
if lowCheck == False:
    lst.append("Lowercase Missing")
if upCheck == False:
    lst.append("Uppercase Missing")
if digCheck == False:
    lst.append("Digit Missing")
if speCheck == False:
    lst.append("Special character Missing")

lst2=lst[:-1:]
if len(lst)==0:
    print("OK")
else:
    for j in lst2:
        print(j, end=", ")
    print(lst[-1])
```

```python
# task 2 (list)
numb=int(input("Please enter how many list: "))
maximum=-999999999999999
empty_lst=[]

for i in range(numb):
    N=input("Please enter list elements using space: ")
    lst1=[int(i) for i in N.split(" ")]
    sumation=sum(lst1)
    if sumation>=maximum:
        maximum=sumation
        empty_lst=lst1
        
print(maximum)
print(empty_lst)
```

```python
# task 4 (list)
while True:
    N=input()
    if N=="STOP":
        break
    else:
        lst=[int(i) for i in N.split(" ")]
        numb=len(lst)
        diff_lst=[]
        for j in range(numb-1):
            a=lst[j+1] - lst[j]
            diff_lst.append(abs(a))
        
        check=True
        for k in range(1,numb):
            if k not in diff_lst:
                check=False
            else:
                continue
        
        if check==True:
            print("UB Jumper")
        else:
            print("Not UB Jumper")
```

```python
# task 5 (list)
N = input("Type here: ")

low_lst=[]
up_lst=[]
odd_lst=[]
eve_lst=[]

for i in N:
    if 97 <= ord(i) <= 122:
        low_lst.append(ord(i))
    elif 65 <= ord(i) <= 90:
        up_lst.append(ord(i))
    elif 48 <= ord(i) <= 57:
        if int(i)%2==0:
            eve_lst.append(ord(i))
        else:
            odd_lst.append(ord(i))

def sort_function(lst):
    new_lst=[]
    while lst:
        min = lst[0]  
        for x in lst: 
            if x < min:
                min = x
        new_lst.append(min)
        lst.remove(min)
    return new_lst

l1=sort_function(low_lst)
l2=sort_function(up_lst)
l3=sort_function(odd_lst)
l4=sort_function(eve_lst)

final_lst=l1+l2+l3+l4

for j in final_lst:
    print(chr(j), end="")
```

**DICTIONARY & TUPLE**

```python
# task 1 (dictionary & tuple)
N1=input("Please enter first dictionary: ")
N2=input("Please enter second dictionary: ")

def dictionary(n):
    dic={}
    lst1 = n.split(",")
    for i in lst1:
        k,v = i.split(":")
        dic[k]=int(v)
    return dic

dic1=dictionary(N1)
dic2=dictionary(N2)

new_dic={}
for k,v in dic2.items():
    if dic1.get(k) == None:
        dic1[k] = v
    else:
        dic1[k] = dic1[k]+v

empty_lst=[]
for a in dic1:
    empty_lst.append(dic1[a])
    
for i in range(len(empty_lst)-1):
    for j in range(0,len(empty_lst)-i-1):
        if empty_lst[j]>empty_lst[j+1]:
            temp=empty_lst[j]
            empty_lst[j]=empty_lst[j+1]
            empty_lst[j+1]=temp

final_lst=[]
for j in empty_lst:
    if j in final_lst:
        continue
    else:
        final_lst.append(j)

print(dic1)
print("Values: ",tuple(final_lst))
```

```python
# task 2 (dictionary & tuple)
lst1=[]
print("Please enter numbers: ")
while True:
    N = input()
    if N=="STOP":
        break
    else:
        lst1.append(N)

dlst=[]
for i in lst1:
    if i not in dlst:
        dlst.append(i)
        
dic1={}
for j in dlst:
    count=0
    for k in lst1:
        if k==j:
            count+=1
    string2="{} times".format(count)
    dic1[int(j)]=string2

for k,v in dic1.items():
    print("{} - {} times".format(k,v))
```

```python
# task 3 (dictionary & tuple)
N=input()
dic={}
lst1 = N.split(",")
for i in lst1:
    k,v = i.split(":")
    dic[k]=v
    
key_lst=[]
value_lst=[]
empty_dic={}
for k,v in dic.items():
    key_lst.append(k)
    value_lst.append(v)
    
for j in range(len(value_lst)):
    if value_lst[j] in empty_dic:
        empty_dic[value_lst[j]].append(key_lst[j])
    else:
        empty_dic[value_lst[j]]=[key_lst[j]]
        
print(empty_dic)
```

```python
# task 4 (dictionary & tuple)
N1=input("Enter first string: ")
N2=input("Enter second string: ")

def dictionary(n):
    dic={}
    for i in n:
        dic[i]=[]
    return dic

dic1=dictionary(N1)
dic2=dictionary(N2)

final_dic={}
count=0
for k1 in dic1:
    if k1 in dic2:
        continue
    else:
        count+=1

if count == 0:
    print("Those strings are anagrams.")
else:
    print("Not anagrams.")
```
